```python
url = 'https://nl.wikipedia.org/wiki/Lijst_van_Radio_2-Top_2000%27s'
df = pd.read_html(url)[0].rename(columns={'Jaar': 'JaarGemaakt'})
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
table = soup.find('table')

artist_links = []
song_links = []
for tr in table.findAll("tr"):
    trs = tr.findAll("td")
    for i, each in enumerate(trs[:2]):
        if i == 0:
            try:
#                 link = each.find('a')['href']
                link = tuple(l['href'] for l in each.find_all('a'))
                artist_links.append(link)
            except:
                pass
        if i == 1:
            try:
                link = each.find('a')['href']
                song_links.append(link)
            except:
                pass
```

```python
df['ArtistLink'] = pd.Series(artist_links)
df['SongLink'] = 'https://nl.wikipedia.org' + pd.Series(song_links)
df['NrArtists'] = df['ArtistLink'].apply(len)
df['ArtistLink'] = df['ArtistLink'].apply(lambda x: 'https://nl.wikipedia.org' + x[0] )
```

```python
def split_top_2000(df):
    df['SongID'] = df[['Titel', 'Artiest']].applymap(lambda x: x.lower()).apply(lambda x: hash(tuple(x)), axis=1)
    songs = df[['SongID', 'Titel', 'Artiest', 'JaarGemaakt', 'ArtistLink', 'SongLink', 'NrArtists']].copy()
    notering = (df[['SongID'] + [col for col in df.columns if col.isnumeric()]]
                .melt(id_vars=['SongID'], var_name='Jaar', value_name='Rank')
                 .assign(Jaar = lambda df: df['Jaar'].astype(int).add(2000).mask(lambda s: s.ge(2050), lambda s: s.sub(100)),
                        Rank = lambda df: pd.to_numeric(df['Rank'], errors='coerce', downcast='integer')
                        )
                .dropna()
                .assign(Rank = lambda df: df['Rank'].astype(int))
               )
    return songs, notering
```

```python
def nr_votes_per_position(position, votes_first_place=25000, diff_factor=30):
    '''
    Source: https://www.petermeindertsma.nl/blog/benadering-aantal-stemmen-per-liedje-in-de-top-2000-van-2014/
    '''
    return votes_first_place / (1 + ( (position - 1) / diff_factor))

def total_votes(nr_songs=5000):
    return sum(nr_votes_per_position(i) for i in range(1, nr_songs))
```

```python
def pct_votes(position):
    return nr_votes_per_position(position)/total_votes()
```

```python
def death_effect(x, minimum_effect=1.5, maximum_effect=3, days_start=30):
    
    return minimum_effect + x.add(days_start).clip(lower=0).div(days_start) * (maximum_effect - minimum_effect)
```

```python
popt, pcov = scipy.optimize.curve_fit(death_effect, effect_overlijden['DaysToStemperiode'], effect_overlijden['OverlijdensBoost'], [1.5, 3, 30])
perr = np.sqrt(np.diag(pcov))
popt
```
